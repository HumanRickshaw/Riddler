# Setup

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (16, 8))
ax = fig.add_subplot(111, xlim = (-1, 1), ylim = (-1, 1))

#Remove axes and ticks.
for side in ['bottom', 'left', 'top', 'right']:
    ax.spines[side].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])

#Triangular Pyramid.
top = [[-0.9, 0.9, 0.3, -0.9], [0.9, 0.9, 0.3, 0.9], 'lightgreen']
left = [[-0.9, 0.3, 0.1, -0.9], [0.9, 0.3, -0.75, 0.9], 'darkgreen']
right = [[0.3, 0.9, 0.1, 0.3], [0.3, 0.9, -0.75, 0.3], 'darkgreen'] 
faces = [top, left, right]
for face in faces:
    ax.plot(face[0], face[1], c = 'black', ls = '-', lw = 2) 
    ax.fill(face[0], face[1], c = face[2], alpha = 0.7)

#Base and Height.
base1 = [[-0.9, 0.1], [0.9, 0.7], 'black', '--']
base2 = [[0.1, 0.6], [0.7, 0.6], 'black', ':']
height = [[0.1, 0.1], [-0.75, 0.7], 'black', '--']
segments = [base1, base2, height] 
for seg in segments :
    ax.plot(seg[0], seg[1], c = seg[2], ls = seg[3], lw = 2)
    
#Lengths.
values = [["1", (0, 0.92), 28],
          ["1", (0.6, 0.5), 28],
          ["1", (-0.3, 0.48), 28],
          ["1/2", (-0.43, -0.2), 28],
          ["1/2", (0.22, -0.2), 28],
          ["1/2", (0.42, -0.2), 28],
          ["1/3", (-0.23, 0.67), 20]]
for value in values :
    plt.annotate(value[0], value[1], fontsize = value[2])
```
