## Setup

```python
import math
from matplotlib.patches import Polygon as poly
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fig = plt.figure(figsize = (11, 11))
ax = fig.add_subplot(111, xlim = (-1.2, 0.85), ylim = (-0.85, 1.2))

#Remove axes and ticks.
for side in ['bottom', 'left', 'top', 'right']:
    ax.spines[side].set_visible(False)
ax.set_xticks([])
ax.set_yticks([])

m = -0.75
n = 0.75

v1 = '#1F968B'
v2 = '#481567'

branches = [[(m, 0, 0, m, m), (n, n, 0, 0, n), v1],
            [(0, n, 0, 0), (n, n, 0, n), v1],
            [(n, n, 0, n), (0, n, 0, 0), v2],
            [(0, n, n, 0, 0), (0, 0, m, m, 0), v2],
            [(0, 0, m, 0), (0, m, m, 0), v2],
            [(m, 0, m, m), (0, 0, m, 0), v1]]

for branch in branches:
    ax.fill(branch[0], branch[1], c = branch[2], alpha = 0.7)
    ax.plot(branch[0], branch[1], c = 'black', ls = '-', lw = 2 ) 
    
rrs = [["First Run", (-1, -0.15), 28, 90],
       ["You are faster", (-0.83, 0.15), 20, 90],
       ["Opponent is faster", (-0.83, -0.68), 20, 90],
       ["Second Run", (-0.27, 1), 28, 0],
       ["You are faster", (-0.6, 0.8), 20, 0],
       ["Opponent is faster", (0.07, 0.8), 20, 0]]

for rr in rrs :
    plt.annotate(rr[0], rr[1], fontsize = rr[2], rotation = rr[3])
```
